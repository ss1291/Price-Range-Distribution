# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read the XLSB file

```python
df = pd.read_excel('C:/Users/admin/Desktop/Cognifyz Internship Projects/dataset.xlsb', engine='pyxlsb')
```

```python
df
```

# Calculate the frequency of each price range category

```python
price_range_counts = df['Price range'].value_counts()
price_range_counts
```

# Calculate the percentage of restaurants in each price range category

```python
total_restaurants = len(df)
price_range_percentages = (price_range_counts / total_restaurants) * 100
print(price_range_percentages)
```

# Create a bar chart to visualize the distribution of price ranges

```python
plt.figure(figsize=(7, 5))
price_range_percentages.plot(kind='bar', color='skyblue')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# Conclusion

The histogram reveals the distribution of price ranges among restaurants, with the majority falling within a Price range 1. The analysis indicates a notable percentage of restaurants concentrated in certain price brackets, shedding light on the pricing dynamics and affordability within the dining landscape.
